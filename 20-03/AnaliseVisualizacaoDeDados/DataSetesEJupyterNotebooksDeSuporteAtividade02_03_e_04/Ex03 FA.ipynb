```python
#Tendo por base o dataset ecommerce_customers.csv e os ficheiros 
#exemplos fornecidos, elabora um notebook jupyter no qual efetues:
#• Limpeza e tratamento de dados;
#• Processamento de dados: groupby, filtering, criação de novas colunas,….;
#• Visualização de dados;
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
sns.set() # seaborn's method to set its chart style
#1º passo importar dados
ecomcust=pd.read_csv("ecommerce_customers.csv")
print(ecomcust.head())
print(ecomcust.tail())
print(ecomcust.info())
print(ecomcust.describe())
print(ecomcust.value_counts())
# view the data  types of each column
print(ecomcust.dtypes)
```

```python
# verificação de valores nulos, neste caso não existem 
ecomcust.info()
print(ecomcust)
```

```python
#Tratar valores duplicados
ecomcust.drop_duplicates()
```

```python
# aqui quis tratar  os dados  tempo na app, tempo no weebsite e montante gasto
 
#vou usar funções min, max e count
ecomcust.groupby(["Time on App","Time on Website","Length of Membership","Yearly Amount Spent"]).size().head()


```

```python
ecomcust.groupby(["Time on App","Time on Website","Length of Membership","Yearly Amount Spent"]).min().head()
```

```python
#A ordenação é feita pelo tempo médio na sessão
#No entanto ordena as colunas definidas do maior para o menor e só depois ordena pelo tempo médio da sessaõ
ecomcust.groupby(["Time on App","Time on Website","Length of Membership","Yearly Amount Spent"]).max().head()
```

```python
ecomcust.groupby(["Time on App","Time on Website","Length of Membership","Yearly Amount Spent"]).count().head()
```

```python
#filtro por Valor gasto maior que 312.30
ecomcust[ecomcust["Yearly Amount Spent"]>312.30].head()
```

```python
#Adicionar a coluna de % que representa tempo na app e total anual gasto
#o comando pd.set_option("display.precision",2) define o numero de casas decimais a apresentar
ecomcust["Tempo na app/Totais gastos (%)"]=(ecomcust["Time on App"]/ecomcust["Yearly Amount Spent"])*100
pd.set_option("display.precision",2)
ecomcust
```

```python
# Criar o plot
fig, (ax0, ax1) = plt.subplots(nrows=2,
ncols=1,
sharex=True,
figsize=(8, 8))
# Adicionar os dados para ax0
scatter = ax0.scatter(ecomcust["Avg. Session Length"],ecomcust
["Yearly Amount Spent"])
# Customizar ax0
ax0.set(title="Avg. Session Length/Yearly Amount Spent",
xlabel="Avg. Session Length",
ylabel="Yearly Amount Spent")
# Adicionar a linha média horizontal para total gasto em ax0
ax0.axhline(ecomcust["Yearly Amount Spent"].mean(),
linestyle="--");
scatter = ax1.scatter(ecomcust["Time on App"],
ecomcust["Length of Membership"])
# Customizar ax1
ax1.set(title="Time on App/Length of Membership",
xlabel="Time on App",
ylabel="Length of Membership")

# Adicionar a linha média para duração de membro
ax1.axhline(ecomcust["Length of Membership"].mean(),
linestyle="--")
# Título da figura
fig.suptitle("Análise de tempo e fidelização", fontsize=16, 
fontweight="bold");
```
